# **IRIS FLOWER CLASSIFICATION**

# About Dataset

The dataset contains four different features of Iris flower. The datas are collected to quantify the morphologic variation of Iris flowers of three related species.The dataset contains 50 samples from each of the related species .Four features were measured from each of the samples : length and width of the petals and sepals in centimeters.

The dataset contains 5 attributes from 150 records : petal length , petal width , sepal length , sepal width , species.

The 3 different related species are 'Iris Setosa','Iris Virginica','Iris Versicolor'.





# Exploratory Data Analysis

```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Dataset/Iris.csv")
df
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
df
```

>> Since 'Id' is not required for further predictions and it is not a necessary feature ,it is dropped.

```python
df.drop('Id',axis=1,inplace=True)
```

```python
df
```

## Separating X and Y

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

#  Scaling & Train Test Split

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

# Model Building

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

```python
knn=KNeighborsClassifier()
sv=SVC()
dtc=DecisionTreeClassifier()
nb=GaussianNB()
rf=RandomForestClassifier()
ab=AdaBoostClassifier()
lst=[knn,sv,dtc,nb,rf,ab]
for i in lst:
  print("-----------",i,"--------------")
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  print(classification_report(y_test,y_pred))
  print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
  y_new=i.predict(scaler.transform([['6.7','3.0','5.2','2.3']]))
  if y_new==0:
    print("Iris-setosa")
  elif y_new==1:
    print("Iris versicolor")
  else:
    print("Iris-virginica")
```

# Visualization

```python
import matplotlib.pyplot as plt
X=[96,96,96,93,96,96]
y=['knn','svc','dtc','nb','rf','ab']
plt.ylabel('Accuracy Score')
plt.xlabel('Classification Algorithm')
plt.title('Accuracy Score vs Classification Algorithm ')
plt.bar(y,X,width=0.5,color="blue")
plt.show()
```
